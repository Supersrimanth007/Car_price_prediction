```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

# Assuming you have already loaded your data into a DataFrame called df
X = df.drop('selling_price', axis=1)
y = df['selling_price']

# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Extracting object-type columns
object_columns = X.select_dtypes(include='object').columns.tolist()

# One-hot encoding categorical variables with handling of unknown categories
ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit(X[object_columns])
column_trans = ColumnTransformer(
    transformers=[('onehot', ohe, object_columns)],
    remainder='passthrough')


# Creating a pipeline with column transformer and linear regression
lr = LinearRegression()
pipe = make_pipeline(column_trans, lr)

# Fitting the pipeline
pipe.fit(X_train, y_train)

# Predicting on the test set
y_pred = pipe.predict(X_test)

# Printing predictions
print(y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
for i in range(10):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans, lr)
    pipe.fit(X_train, y_train)
    y_pred=pipe.predict(X_test)
    print(r2_score(y_test, y_pred), i)
```
